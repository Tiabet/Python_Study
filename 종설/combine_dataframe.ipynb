```python
import pandas as pd
```

```python
df_books = pd.read_csv('books_newsdata_embedding.csv')
```

```python
df_news = pd.read_csv('news_newsdata_embedding.csv')
```

```python
df_news.head(5)
```

```python
df_books.head(5)
```

```python
df_books = df_books[['id', 'url', 'published_date', 'title', 'embedding']]
```

```python
df = pd.concat([df_news,df_books])
```

```python
df['published_date'] = pd.to_datetime(df['published_date'])
```

```python
df
```

```python
# Sort by date
sorted_df = df.sort_values(by='published_date')
```

```python
sorted_df.drop('id',axis=1,inplace=True)
```

```python
sorted_df.to_csv('newsdata_embedding.csv')
```

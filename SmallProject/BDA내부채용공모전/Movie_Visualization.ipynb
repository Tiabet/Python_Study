```python
import pandas as pd

df = pd.read_csv("영화_최종데이터_v3.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Success', hue='Success', palette=['#4c72b0', '#55a868'])
plt.title('Count of Movies by Success')
plt.xlabel('Success (0 or 1)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend([],[], frameon=False)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='genre1', palette='viridis')
plt.title('Count of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='genre1', hue='Success', palette='viridis')
plt.title('Count of Movies by Genre and Success')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

selected_columns_for_analysis = [
    'sentiment','rating','ViewCount', 'LikeCount', 'Actor1', 'Actor2', 'Actor3', 'GENRE1', 'GENRE2',
    '남자 인구수', '여자 인구수', '총인구수', '가계지출', '소비지출', '비소비지출', 'Total Sales'
]

data_filled = df[selected_columns_for_analysis].fillna(df[selected_columns_for_analysis].mean())

correlation_matrix_filtered = data_filled.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_filtered, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix with Success Metrics')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='Production Budgets', kde=True, color='skyblue')
plt.title('Distribution of Production Budgets')
plt.xlabel('Production Budgets')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='Success', y='Production Budgets', palette='viridis')
plt.title('Success vs Production Budgets')
plt.xlabel('Success')
plt.ylabel('Production Budgets')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.catplot(data=df, x='Month', hue='Success', col='genre1', kind='count', palette='viridis', height=4, aspect=0.8, col_wrap=4)
plt.suptitle('Count of Movies by Month, Success, and Genre', y=1.02)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("영화_최종데이터_v4.csv")
```

```python
df.columns
```

### 관람연령 인코딩

```python
df['Level'].unique()

category_mapping = {'전체관람가': 1, '12세이상관람가': 2, '15세이상관람가': 3, '청소년관람불가': 4}
df['LEVEL'] = df['Level'].replace(category_mapping)
df
```

### 배우1,2,3, 장르1,2 평균값 구해서 하나의 열로 만들기['Actor','Genre']

```python
df['Actor'] = df[['Actor1','Actor2','Actor3']].mean(axis = 1,skipna=True).round(2)
df['Genre'] = df[['GENRE1','GENRE2']].mean(axis = 1,skipna=True).round(2)
```

```python
ddf = df[['Genre','DIRECTOR','LEVEL','Actor','sentiment','ViewCount','LikeCount','PRODUCTION COMPANY']]
```

### 평균값으로 null 값 대체

```python
mean_values = ddf.mean()
ddf_filled = ddf.fillna(mean_values)
ddf_filled
```

### 스케일링

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
scaler = StandardScaler()

# 데이터프레임의 각 열에 대해 StandardScaler 적용
scaled_data = scaler.fit_transform(ddf_filled)

# 스케일링된 데이터로 새로운 데이터프레임 생성
df_scaled = pd.DataFrame(scaled_data, columns=ddf_filled.columns)
```

```python
# null 값이 있는 행 찾기
rows_with_null = df_scaled[df_scaled.isnull().any(axis=1)]

print("Null 값이 있는 행:")
rows_with_null
```

### modeling

```python
df_x = df_scaled[['Genre','DIRECTOR','LEVEL','Actor','sentiment','ViewCount','LikeCount','PRODUCTION COMPANY']]
df_y = df['Success']
```

### 로지스틱

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier



# 학습용과 검증용 데이터 분할
df_x_train, df_x_test, df_y_train, df_y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)

# 로지스틱 회귀 모델 학습
model = LogisticRegression()
model.fit(df_x_train, df_y_train)

# 검증용 데이터로 예측
y_pred = model.predict(df_x_test)

# 모델 평가
accuracy = accuracy_score(df_y_test, y_pred)
print(f'정확도: {accuracy:.2f}')

report = classification_report(df_y_test, y_pred)
print('분류 보고서:\n', report)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report


# 수치형 데이터는 스케일링
numerical_cols = [ 'DIRECTOR', 'Actor', 'PRODUCTION COMPANY', 'sentiment', 'ViewCount', 'LikeCount']
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
```

```python
# 전처리 파이프라인 정의
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols)
    ])

# 모델 정의: 랜덤 포레스트 분류기
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 파이프라인 정의
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', model)])

# 교차 검증 (Cross-validation)을 통한 모델 평가
cv_scores = cross_val_score(clf, df_x, df_y, cv=5, scoring='accuracy')

print(f"평균 정확도 (Cross-validation 평균): {cv_scores.mean():.2f}")
print("각 Fold의 정확도:", cv_scores)

# 전체 데이터를 사용하여 모델을 학습시킴 (실제 예측시에는 학습된 모델을 사용하여 예측)
clf.fit(df_x, df_y)

# 예측 및 성능 평가
y_pred = clf.predict(df_x)
print("\n전체 데이터에 대한 분류 보고서:\n", classification_report(df_y, y_pred))
```

```python
#최종
#url_list = ['http://cine21.com/movie/info/?movie_id=5489','http://cine21.com/movie/info/?movie_id=61313', 'http://cine21.com/movie/info/?movie_id=9510','http://cine21.com/movie/info/?movie_id=59264']
browser = webdriver.Chrome()
review_data = pd.DataFrame()

for u in final_listt:
    
    browser.get(u)  # 대상 페이지 URL 입력
    browser.page_source
    time.sleep(3)  # 페이지 로드 대기
    soup = BeautifulSoup(browser.page_source, 'html.parser', from_encoding='utf-8')
    review_area = soup.find('div', id='netizen_review_area')
    if review_area:
        review_text = review_area.find('h2').find('span').text
        review_count = re.search(r'\d+', review_text).group()
        print("네티즌 리뷰 수:", review_count)
    else:
        print("네티즌 리뷰 수를 찾을 수 없습니다.")

    # 영화 이름
    movie_title_element = soup.find('p', class_='tit')
    movie_title = movie_title_element.text.strip() if movie_title_element else ''
    print("영화 제목:", movie_title)

    # 상영시간
    time_element = soup.find('span', text=lambda x: x and '상영시간' in x)
    running_time = time_element.text.split(':')[1].strip() if time_element else ''
    print("상영시간:", running_time)
    
    # 개봉일
    release_date_element = soup.find('span', text=lambda x: x and '개봉일' in x)
    release_date = release_date_element.text.split(':')[1].strip() if release_date_element else ''
    print("개봉일:", release_date)

    # 누적관객
    viewer_element = soup.find('span', text=lambda x: x and '누적관객' in x)
    viewer = viewer_element.text.split(':')[1].strip() if viewer_element else ''
    print("누적관객:", viewer)
    
    # 감독
    director = ''
    director_section = soup.find('h3', text='감독')
    if director_section:
        director_list = director_section.find_next('ul', class_='list_big').find_all('p', class_='name')
        for item in director_list:
            director += item.text.strip() + ', '
    print(director)
    


    actors = []
    actors_section = soup.find('h3', text='출연')
    if actors_section:
        ul_element = actors_section.find_next('ul', class_='list_big')
        if ul_element:
            actor_list = ul_element.find_all('li')
            for actor_item in actor_list:
                actor_name = actor_item.find('p', class_='name').text.strip()
                actors.append(actor_name)
                # Process actor_name
        else:
            print("No <ul> element with class 'list_big' found after '출연' section.")
    else:
        print("No <h3> element with text '출연' found.")

    
    
    # 제작 정보 추출
    production = None
    production_element = soup.find('p', class_='tit', text='제작')
    if production_element:
        production = production_element.find_next_sibling('a').text.strip()
    print("제작:", production if production else '')

    
    
    # 각본 작가
    writers = []
    writers_section = soup.find('p', class_='tit', text='각본')
    if writers_section:
        writer_list = writers_section.find_next_siblings('a')
        for item in writer_list:
            writers.append(item.text.strip())
    print("각본 작가:", ', '.join(writers))

    # 촬영 감독
    cinematographers = []
    cinematographers_section = soup.find('p', class_='tit', text='촬영')
    if cinematographers_section:
        cinematographer_list = cinematographers_section.find_next_siblings('a')
        for item in cinematographer_list:
            cinematographers.append(item.text.strip())
    print("촬영 감독:", ', '.join(cinematographers))

    # 편집자
    editors = []
    editors_section = soup.find('p', class_='tit', text='편집')
    if editors_section:
        editor_list = editors_section.find_next_siblings('a')
        for item in editor_list:
            editors.append(item.text.strip())
    print("편집자:", ', '.join(editors))

    # 원작
    original_works = []
    original_section = soup.find('p', class_='tit', text='원작')
    if original_section:
        original_list = original_section.find_next_siblings('a')
        for item in original_list:
            original_works.append(item.text.strip())
    print("원작:", ', '.join(original_works))

    # 배급사
    distributors = []
    distributors_section = soup.find('p', class_='tit', text='배급')
    if distributors_section:
        distributor_list = distributors_section.find_next_siblings('a')
        for item in distributor_list:
            distributors.append(item.text.strip())
    print("배급사:", ', '.join(distributors))

    # 수입사
    importers = []
    importers_section = soup.find('p', class_='tit', text='수입')
    if importers_section:
        importer_list = importers_section.find_next_siblings('a')
        for item in importer_list:
            importers.append(item.text.strip())
    print("수입사:", ', '.join(importers))

    button_num = ['div/a[1]','div/a[2]', 'div/a[3]','div/a[4]','div/a[5]','div/a[6]','div/a[7]','div/a[8]','div/a[9]','div/a[10]']
    next_button_path = '//*[@id="netizen_review_area"]/div/a[2]/span[2]'
    

    if int(review_count)>0:
        while True:
            data_list = []
            try:
                for i in button_num:
                    
                    page_button = browser.find_elements(By.XPATH,'//*[@id="netizen_review_area"]/div/'+i)
                    print(page_button)
                    wait = WebDriverWait(browser, 10)
                    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="netizen_review_area"]/div/'+i)))
                    element.click()
                    time.sleep(3)  # 페이지 로드 대기
                     # 페이지 내 리뷰 추출
                    review_items = soup.select('.reply_box li')
                    for a in review_items:
                        review_id = a.find('div', class_='id').text.strip()
                        review_date = a.find('div', class_='date').text.strip()
                        review_rating = a.find('span', class_='num').text.strip()
                        review_comment = a.find('div', class_='comment').text.strip()
                        
                        data = {
                            "movie_title" : movie_title,
                                "running_time": running_time,
                                "release_date": release_date,
                                "viewer" : viewer,
                                "director" : director,
                                "actors" : actors,
                                "production" : production, 
                                "screenplay_writers": writers ,
                                "director_of_photography": cinematographers,
                                "editors": editors, 
                                "original_author":original_works, 
                                "distributor":distributors, 
                                "importer":importers,
                            "review_count": review_count,
                            
                          
                            "review_id": review_id,
                            "review_date": review_date,
                            "review_rating": review_rating,
                            "review_comment": review_comment
                        }
                        data_list.append(data)
            except Exception as e_a:
                break # 한 페이지에서 다음 페이지가 없을 때 혹은 페이지 넘기기 버튼을 눌러야 할 때
        
            try:
                next_button = browser.find_elements(By.XPATH, next_button_path) # 첫 넘기기 버튼 누기기
                next_button[0].click()
                time.sleep(3)  # 페이지 로드 대기
            
                next_button_path = '//*[@id="netizen_review_area"]/div/a[3]/span[2]' # 넘기기 버튼 x_path 갱신
            
            except Exeption as e_b:
                break
                
    elif int(review_count) == 0 :
            data_list = []
            data = {
                                "movie_title" : movie_title,
                                "running_time": running_time,
                                "release_date": release_date,
                                "viewer" : viewer,
                                "director" : director,
                                "actors" : actors,
                                "production" : production, 
                                "screenplay_writers": writers ,
                                "director_of_photography": cinematographers,
                                "editors": editors, 
                                "original_author":original_works, 
                                "distributor":distributors, 
                                "importer":importers, 
                                "review_count": review_count,
                                "review_id": '',
                                "review_date": '',
                                "review_rating": '',
                                "review_comment": ''}
            data_list.append(data)    

    df = pd.DataFrame(data_list)
    review_data = pd.concat([review_data, df], ignore_index=True)



# 결과 출력
print(review_data)

            
```

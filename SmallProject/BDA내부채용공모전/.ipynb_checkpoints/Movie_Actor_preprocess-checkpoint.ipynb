### 배우 누적 관객수 평균값 치환

```python
actor1 = df.groupby('actor1')['Cumulative Attendance'].mean()

actor1.dropna()
```

```python
df.groupby('actor2')['Cumulative Attendance'].mean()
```

```python
df.groupby('actor3')['Cumulative Attendance'].mean()
```

```python
# actor1, actor2, actor3 열에서 중복 제거한 데이터 가져오기
unique_actors = pd.concat([df['actor1'], df['actor2'], df['actor3']]).drop_duplicates().reset_index(drop=True)
```

```python
unique_actors.drop(8,inplace=True)
```

```python
unique_actors
```

```python
actors_cum_view_mean = []
for a in unique_actors:

    total = 0 
    row_count = 0
    for i in range(0,len(df)):
        if df['actor1'][i] == a or df['actor2'][i] == a or df['actor3'][i] == a:
            total += int(df['Cumulative Attendance'][i])
            row_count += 1
    mean_value = round(total/ row_count,1)
    actors_cum_view_mean.append(mean_value)
    print(a)
    print(total)
    print(row_count)
    print(mean_value)
    print('-----------')
    
        
            
            


    
```

```python
actor_fam = pd.DataFrame()

actor_fam['actor'] = unique_actors
actor_fam['mean'] = actors_cum_view_mean

actor_mean_dict = dict(zip(actor_fam['actor'], actor_fam['mean']))
```

```python
actor_mean_dict
```

```python
# 치환 할 열 가가

df['Actor1'] = df['actor1']
df['Actor2'] = df['actor2']
df['Actor3'] = df['actor3']
```

```python
# 배우 별 관객수 평균값 치환
column = ['Actor1','Actor2','Actor3']

for col in column:
    df[col] = df[col].map(actor_mean_dict)
```

```python
df.to_csv('배우_전처리완료.csv')
```

### 개봉 월 & 장르 

```python
from datetime import datetime

df['Month'] = df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d').month)
```

```python
df.columns
```

```python
genre = pd.concat([df['genre1'], df['genre2']]).drop_duplicates().reset_index(drop=True)
```

```python
result = df.groupby(['genre1','Month'])['genre1'].count().reset_index(name='count')
pd.set_option('display.max_rows', None) 
result
```

```python
result2 = df.groupby(['genre2','Month'])['genre2'].count().reset_index(name='count')
pd.set_option('display.max_rows', None) 
result2
```

```python
import pandas as pd


# genre1과 genre2를 합친 열 추가
merged_df = pd.merge(result, result2, left_on=['genre1', 'Month'], right_on=['genre2', 'Month'], how='outer')

# 필요한 열들만 선택하여 새로운 열 생성
merged_df['genre'] = merged_df['genre1'].fillna(merged_df['genre2'])
merged_df = merged_df[['genre', 'Month', 'count_x', 'count_y']]

# count 값 합계 계산
merged_df['count'] = merged_df['count_x'].fillna(0) + merged_df['count_y'].fillna(0)

# 필요없는 열 제거
merged_df = merged_df[['genre', 'Month', 'count']]

print("Merged DataFrame:")
print(merged_df)
```

```python
# 각 장르별 월별 count의 합 구하기
genre_month_sum = merged_df.groupby(['genre', 'Month'])['count'].sum().reset_index()

# 각 장르별 전체 count 합 구하기
genre_total_sum = merged_df.groupby('genre')['count'].sum()

# 장르별 월별 비율 계산
genre_month_sum['ratio'] = genre_month_sum.apply(lambda row: round(row['count'] / genre_total_sum[row['genre']],2), axis=1)

print("장르별 월별 비율:")
print(genre_month_sum)
```

```python
genre_month_df = genre_month_sum.sort_values(['genre','Month'])
genre_month_df
```

```python
def replace_with_ratio(row, ratio_df, column_name):
    genre = row[column_name]
    month = row['Month']
    ratio_value = ratio_df[(ratio_df['genre'] == genre) & (ratio_df['Month'] == month)]['ratio'].values
    if len(ratio_value) > 0:
        return ratio_value[0]
    else:
        return None
```

```python
# genre1 열의 값을 비율로 치환하여 GENRE1 열에 추가
df['GENRE1'] = df.apply(lambda row: replace_with_ratio(row, genre_month_df, 'genre1'), axis=1)

# genre2 열의 값을 비율로 치환하여 GENRE2 열에 추가
df['GENRE2'] = df.apply(lambda row: replace_with_ratio(row, genre_month_df, 'genre2'), axis=1)

print("Updated DataFrame:")
df[['genre1','genre2','Month','GENRE1','GENRE2']]
```

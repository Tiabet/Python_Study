```python
additional_df = pd.read_parquet('newsdata_addtional.parquet')
```

```python
original_df = pd.read_parquet('newsdata_preprocessed+embedding.parquet')
```

```python
original_df.rename(columns = {'published_date' : 'date'}, inplace = True)
```

```python
additional_df['date'][0]
```

```python
original_df['date'] = pd.to_datetime(original_df['date'], format='%Y-%m-%d')
```

```python
final_df = pd.concat([additional_df, original_df])
```

```python
final_df = final_df.reset_index(drop = True)
```

```python
final_df
```

```python
final_df.sort_values(by = 'date')
```

```python
def text_preprocess(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    
    return text
```

```python
final_df['title'] = final_df['title'].apply(text_preprocess)
```

```python
final_df.drop('embedding', axis = 1, inplace = True)
```

```python
final_df.to_parquet('newsdata.parquet', engine = 'pyarrow', compression = 'gzip', index = False)
```

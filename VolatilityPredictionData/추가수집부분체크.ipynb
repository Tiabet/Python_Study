```python
original_df = pd.read_csv("1차정제데이터.csv")
```

```python
additional_df = pd.read_csv("결측치수집데이터.csv")
#additional_df2 = pd.read_csv("결측치수집데이터2.csv")
```

```python
# Find duplicated dates in the original dataframe
duplicated_dates = original_df['date']

# Filter additional dataframe to remove rows with duplicated dates
filtered_additional_df = additional_df[~additional_df['date'].isin(duplicated_dates)]
#filtered_additional_df2 = additional_df2[~additional_df2['date'].isin(duplicated_dates)]
```

```python
filtered_additional_df = filtered_additional_df.reset_index()
#filtered_additional_df2 = filtered_additional_df2.reset_index()
```

```python
#filtered_additional_df.drop('index', axis = 1, inplace = True)
filtered_additional_df = filtered_additional_df.sort_values(by = 'date')
#filtered_additional_df2.drop('index', axis = 1, inplace = True)
#filtered_additional_df2 = filtered_additional_df2.sort_values(by = 'date')
```

```python
filtered_additional_df.to_csv("처리해야할파트1.csv",index=False)
```

```python
#filtered_additional_df2.to_csv("처리해야할파트2.csv",index=False)
```

```python
#new_df = pd.concat([filtered_additional_df2,filtered_additional_df,original_df])
new_df = pd.concat([filtered_additional_df,original_df])
```

```python
start_date = '1986-01-01'
end_date = '2001-12-31'
all_dates = pd.date_range(start=start_date, end=end_date, freq='D')
missing_dates = all_dates.difference(new_df['date'])
missing_df = pd.DataFrame(missing_dates, columns=['date'])
missing_df['year'] = missing_df['date'].dt.year
missing_df['month'] = missing_df['date'].dt.month
missing_counts_per_month = missing_df.groupby(['year', 'month']).size()
print("Missing days per month:")
print(missing_counts_per_month)
sorted_counts = missing_counts_per_month.sort_index()

# Convert the sorted series to a DataFrame
missing_counts_df = sorted_counts.reset_index(name='missing_days')

print(missing_counts_df)
missing_counts_df.sort_values(by = 'missing_days')
print(missing_counts_df)
```

```python
missing_counts_df_15 = missing_counts_df[missing_counts_df['missing_days']>=15]
```

```python
import re

# Example text with irregular white spaces
text = "The room     temperature   is  23.5    degrees, and there   are 15    people inside.  "

# Use re.sub to replace multiple whitespace characters with a single space
text = re.sub(r'\s+', ' ', text)

# Print the processed text
print(text)
```

```python
df1 = pd.read_csv("내담당추가부분.csv")
df2 = pd.read_csv("동연.csv")
df3 = pd.read_csv("주환.csv")
```

```python
df1.drop('Unnamed: 2', axis = 1, inplace = True)
```

```python
df1 = df1.dropna()
```

```python
df2
```

```python
df2['date'] = pd.to_datetime(df2['date'], format = '%Y.%m.%d')
```

```python
additional_df = pd.concat([df1, df2, df3])
```

```python
additional_df = additional_df.reset_index()
additional_df['date'] = pd.to_datetime(additional_df['date'])
```

```python
additional_df.drop('index',axis = 1 ,inplace = True)
```

```python
additional_df = additional_df.sort_values(by = 'date')
```

```python
def text_preprocess(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)

    return text
```

```python
additional_df['title'] = additional_df['title'].apply(text_preprocess)
```

```python
#additional_df = additional_df.reset_index()
additional_df.drop('index',axis = 1 ,inplace = True)
additional_df
```

```python
original_df = pd.read_parquet('1차수집데이터.parquet')
```

```python
final_df = pd.concat([original_df, additional_df])
```

```python
final_df = final_df.groupby('date')['title'].agg(lambda x: ', '.join(x)).reset_index()
```

```python
final_df['title'] = final_df['title'].apply(text_preprocess)
```

```python
final_df = final_df.sort_values(by = 'date')
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# Load the SentenceTransformer model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Function to get embeddings
def get_embeddings(text):
    return model.encode(text)

# Apply the function to the 'title' column
final_df['embeddings'] = final_df['title'].apply(get_embeddings)

#print(df['embeddings'].iloc[0])  # Print the embedding of the first title
```

```python
final_df.to_parquet('크루드오일최종.parquet',
                                 engine = 'pyarrow',
                                 compression = 'gzip'
                                 )
```

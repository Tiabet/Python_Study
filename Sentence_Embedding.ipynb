```python
import pandas as pd
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
df_books = pd.read_csv("/content/drive/MyDrive/books_newsdata.csv")
df_news = pd.read_csv("/content/drive/MyDrive/news_newsdata.csv")

print(len(df_books))
print(len(df_news))
```

```python
import re

def remove_pattern(text):
    pattern = r"- \d+페이지"
    cleaned_text = re.sub(pattern, "", text)
    return cleaned_text
```

```python
df_books['Title'] = df_books['title'].apply(remove_pattern)
```

```python
df_books.drop('title',axis=1,inplace=True)
```

```python
df_books.rename(columns = {'Title' : 'title'},inplace=True)
```

```python
df_books.to_csv('/content/drive/MyDrive/books_newsdata.csv',index=False)
```

```python
# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')

def get_embedding(sentence):
    # Encode the sentence
    encoded_input = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True, max_length=128)
    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input)
    # Perform pooling. For simplicity, we'll just use mean pooling.
    sentence_embedding = torch.mean(model_output.last_hidden_state, dim=1).squeeze()
    return sentence_embedding.numpy()

# Apply the function to get embeddings
df_news['embedding'] = df_news['title'].apply(lambda x: get_embedding(x))
```

```python
# Apply the function to get embeddings
df_books['embedding'] = df_books['title'].apply(lambda x: get_embedding(x))
```

```python
df_news.to_csv('/content/drive/MyDrive/news_newsdata_embedding.csv',index=False)
df_books.to_csv('/content/drive/MyDrive/books_newsdata_embedding.csv',index=False)
```
